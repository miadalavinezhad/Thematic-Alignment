```python
from thematic_alignment import fetch_papers
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
API_KEY = os.getenv("SEMANTIC_SCHOLAR_API_KEY")
```

```python
query = "plos computational biology"
fields = ["title", "abstract", "year", "venue"]
if API_KEY:
    headers = {"x-api-key": API_KEY}
else:
    # Without an API key, there are more restrictions from Semantic Scholar
    headers = {}
# Semantic Scholar API endpoint for paper search
url = "http://api.semanticscholar.org/graph/v1/paper/search"

raw_papers_df = fetch_papers(query=query, fields=fields, headers=headers, url=url)
```
